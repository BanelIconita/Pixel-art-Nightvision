```python
def photo2pixelart(image, i_size, o_size):
    """
    image: path to image file
    i_size: size of the small image eg:(8,8)
    o_size: output size eg:(10000,10000)
    """
    
    #read file
    img=Image.open(image)

    #convert to small image
    small_img=img.resize(i_size,Image.BILINEAR)

    #resize to output size
    res=small_img.resize(img.size, Image.NEAREST)

    res = ImageOps.grayscale(res)
    # red, green, blue = res.split()
    # zeroed_band = red.point(lambda _: 0)
    # res = Image.merge( "RGB", (zeroed_band, green, zeroed_band))
    # enhancer = ImageEnhance.Brightness(res)
    # res = enhancer.enhance(1)

    #Save output image
    filename=f'output_{i_size[0]}x{i_size[1]}.png'
    res.save(filename)

    #Display images side by side
    plt.figure(figsize=(16,10))
    #original image
    plt.subplot(1,2,1)
    plt.title('Original image', size=18)
    plt.imshow(img)   #display image
    plt.axis('off')   #hide axis
    #pixel art
    plt.subplot(1,2,2)
    plt.title(f'Pixel Art {i_size[0]}x{i_size[1]}', size=18)
    plt.imshow(res)
    plt.axis('off')
    plt.show()
```

```python
pixeled_image = photo2pixelart(image='input.png',i_size=(64,64),
                    o_size=img.size)
```
